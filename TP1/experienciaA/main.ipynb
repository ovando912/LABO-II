```python
deltap_mean = np.array(deltap_1_mean + deltap_2_mean + deltap_3_mean)
deltap_mean_error = np.array(
    deltap_1_mean_error + deltap_2_mean_error + deltap_3_mean_error
)
caudal_mean = np.array(caudal_1_mean + caudal_2_mean + caudal_3_mean)
caudal_mean_error = np.array(
    caudal_1_mean_error + caudal_2_mean_error + caudal_3_mean_error
)

plt.figure(figsize=(10, 5))
plt.errorbar(
    caudal_mean,
    deltap_mean,
    xerr=caudal_mean_error,
    yerr=deltap_mean_error,
    fmt="o",
    markersize=3,
    label="Mediciones",
)
plt.xlabel("Caudal Primario")
plt.ylabel("DeltaP Nucleo")
plt.title("Relación entre Caudal Primario y DeltaP Nucleo")
plt.legend()
plt.grid()
plt.show()

# caudal_mean_squared = np.array(caudal_mean) ** 2
caudal_mean_squared = caudal_mean**2
caudal_mean_squared_error = 2 * caudal_mean * caudal_mean_error

import scipy.odr as odr


# Definir el modelo lineal
def linear_func(B, x):
    return B[0] + B[1] * x  # B[0] = a, B[1] = b


# Configurar el modelo ODR
model = odr.Model(linear_func)
data = odr.RealData(
    caudal_mean_squared, deltap_mean, sx=caudal_mean_squared_error, sy=deltap_mean_error
)
odr_instance = odr.ODR(data, model, beta0=[0, 1])  # Valores iniciales [a, b]

# Ejecutar la regresión
output = odr_instance.run()

# Obtener parámetros ajustados y sus errores
a, b = output.beta
a_err, b_err = output.sd_beta

# Mostrar resultados
print(f"a = {a:.4f} ± {a_err:.4f}")
print(f"b = {b:.4f} ± {b_err:.4f}")

# Graficar los datos con la recta ajustada
x_fit = np.linspace(min(caudal_mean_squared), max(caudal_mean_squared), 100)
y_fit = a + b * x_fit

plt.rcParams.update({
    "text.usetex": True,  # Usa LaTeX para el texto
    "font.family": "serif",  # Fuente tipo serif (similar a LaTeX)
    "font.serif": ["Computer Modern Roman"],  # Usa Computer Modern
    "axes.labelsize": 16,  # Tamaño de etiquetas de ejes
    "font.size": 16,  # Tamaño general de fuente
    "legend.fontsize": 16,  # Tamaño de fuente en la leyenda
    "xtick.labelsize": 14,  # Tamaño de fuente en los ticks de x
    "ytick.labelsize": 14,  # Tamaño de fuente en los ticks de y
})


plt.figure(figsize=(10, 5))
# plt.plot(caudal_mean_squared, deltap_mean, "o", label="Mediciones")
plt.errorbar(
    caudal_mean_squared,
    deltap_mean,
    xerr=caudal_mean_squared_error,
    yerr=deltap_mean_error,
    fmt="o",
    markersize=3,
    label="Mediciones",
)
plt.plot(
    x_fit,
    y_fit,
    label=f"Regresión lineal",
    color="red",
)
plt.xlabel("$Q_P^2$ [m$^6$/h$^2$]")
plt.ylabel("$\Delta P_N$ [mm$_{H_2O}$]")
# plt.title(" vs. DeltaP NucleoCaudal Primario")
plt.legend()
plt.grid()
plt.show()
```

```python
data_cruda = {
    '$\Delta P_N$ [mm$_{H_2O}$]': deltap_mean,
    '$\Delta P_N$ error': deltap_mean_error,
    '$Q_P$ [m$^6$/h$^2$]': caudal_mean,
    '$Q_P$ error': caudal_mean_error
}
df_crudo = pd.DataFrame(data_cruda)

def round_to_significant_figures(x, sig_figs):
    return np.round(x, sig_figs - int(np.floor(np.log10(np.abs(x)))) - 1)

# Redondear los valores de error a 1 cifra significativa
deltap_mean_error_rounded = np.array([round_to_significant_figures(err, 1) for err in deltap_mean_error])
caudal_mean_error_rounded = np.array([round_to_significant_figures(err, 1) for err in caudal_mean_error])

def get_decimal_places(x):
    """Devuelve el número de decimales significativos en el número x"""
    if x == int(x):
        return 0
    return len(str(x).split('.')[1])

def round_to_error_precision(value, error):
    """Redondea el valor a la precisión del error"""
    # Obtener el número de decimales en el error
    error_precision = get_decimal_places(error)
    
    # Redondear el valor a la precisión del error
    rounded_value = np.round(value, error_precision)
    return rounded_value

deltap_mean_truncated = np.array([round_to_error_precision(val, err) for val, err in zip(deltap_mean, deltap_mean_error)])
caudal_mean_truncated = np.array([round_to_error_precision(val, err) for val, err in zip(caudal_mean, caudal_mean_error)])
                                 
data = {
    '$\Delta P_N$ [mm$_{H_2O}$]': deltap_mean_truncated,
    '$\Delta P_N$ error': deltap_mean_error_rounded,
    '$Q_P$ [m$^6$/h$^2$]': caudal_mean_truncated,
    '$Q_P$ error': caudal_mean_error_rounded
}
df = pd.DataFrame(data)
```
