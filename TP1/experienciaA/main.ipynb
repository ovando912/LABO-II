```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

# Cargar el archivo CSV (o TXT) en un DataFrame
df = pd.read_csv("DatosBombaPrimario.TXT", delimiter=",")

# Seleccionar las últimas dos columnas
df = df.iloc[:, 1:]  # Esto elimina la columna de 'Hora'

# Grafico los datos
plt.figure(figsize=(10, 5))
plt.plot(df["DPN"], label="DPN")
plt.plot(df["FPM"], label="FPM")
plt.legend()
plt.show()


```

```python
# Separo las 3 mediciones en 3 DataFrames distintos
df1 = df.iloc[1500:2800]
df1 = df1.reset_index(drop=True)
df2 = df.iloc[3400:4100]
df2 = df2.reset_index(drop=True)
df3 = df.iloc[5000:5900]
df3 = df3.reset_index(drop=True)
```

```python
# Grafico los datos
# plt.figure(figsize=(10, 5))
# plt.plot(df1["DPN"], label="DeltaP Nucleo")
# plt.plot(df1["FPM"], label="Caudal Primario")
# plt.title("Medición 1")
# plt.legend()
# plt.show()

medicion_1_beg = [989, 1072, 1166]
medicion_1_end = [1062, 1156, 1259]

deltap_1_mean = []
deltap_1_mean_error = []
caudal_1_mean = []
caudal_1_mean_error = []
for beg, end in zip(medicion_1_beg, medicion_1_end):
    # deltap_aux = df1["DPN"].iloc[beg : end + 1].mean()
    # deltap_aux_error = np.sqrt(
    #     np.sum((df1["DPN"].iloc[beg : end + 1] - deltap_aux) ** 2) / (end - beg)
    # ) / np.sqrt(end - beg + 1)
    # deltap_1_mean.append(deltap_aux)
    # deltap_1_mean_error.append(deltap_aux_error)
    datos = df1["DPN"].iloc[beg : end + 1].to_numpy()
    deltap_aux = np.mean(datos)
    deltap_aux_error = np.std(datos, ddof=1) / np.sqrt(len(datos)) 
    deltap_1_mean.append(deltap_aux)
    deltap_1_mean_error.append(deltap_aux_error)

    # caudal_aux = df1["FPM"].iloc[beg : end + 1].mean()
    # caudal_aux_error = np.sqrt(
    #     np.sum((df1["FPM"].iloc[beg : end + 1] - caudal_aux) ** 2) / (end - beg)
    # ) / np.sqrt(end - beg + 1)
    # caudal_1_mean.append(caudal_aux)
    # caudal_1_mean_error.append(caudal_aux_error)
    datos = df1["FPM"].iloc[beg : end + 1].to_numpy()
    caudal_aux = np.mean(datos)
    caudal_aux_error = np.std(datos, ddof=1) / np.sqrt(len(datos))
    caudal_1_mean.append(caudal_aux)
    caudal_1_mean_error.append(caudal_aux_error)

# plot los datos y los promedios en sus intervalos como segmentos horizontales
plt.figure(figsize=(10, 5))
plt.plot(df1["DPN"], label="DeltaP Nucleo")
plt.plot(df1["FPM"], label="Caudal Primario")
plt.title("Medición 1")
plt.legend()
for beg, end, deltap, caudal in zip(
    medicion_1_beg, medicion_1_end, deltap_1_mean, caudal_1_mean
):
    plt.hlines(deltap, beg, end, color="red")
    plt.hlines(caudal, beg, end, color="green")
plt.show()
```
